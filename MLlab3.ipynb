```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = { "score" :[480,490,510,510,530,550,610,640],
            "gpa":[2.7,2.9,3.3,2.9,3.1,3.0,3.2,3.7]
           }
```

```python
dataset
```

```python
dataframe = pd.DataFrame(dataset)
dataframe
```

```python
X = dataframe['score'].values
y = dataframe['gpa'].values
```

```python
X=X.reshape(-1,1)
```

```python
# Splitting the dataset into Training and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

# Printing the training and test sets (Optional)
print("X_train:\n", X_train)
print("y_train:\n", y_train)
print("X_test:\n", X_test)
print("y_test:\n", y_test)
```

```python
# Training the Simple Linear Regression Model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Getting the Coefficient and Intercept
coefficient = regressor.coef_
intercept = regressor.intercept_

print("Coefficient (Slope):", coefficient[0])
print("Intercept:", intercept)
```

```python
regressor.predict(([600],[700]))
```
